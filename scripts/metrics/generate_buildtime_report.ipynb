```python
import altair as alt # https://altair-viz.github.io/user_guide/interactions.html
import datapane as dp # https://www.youtube.com/watch?v=_KS_yZBI71s&ab_channel=Datapane
import pandas as pd
```

```python
buildtimes_df = pd.read_csv("new_buildtimes.csv")
buildtimes_df
```

```python
dp.DataTable(buildtimes_df)
```

```python
section_df = buildtimes_df.query("section_type == 'subsection' and section_name == 'build workspace in isolation'")
section_df
```

```python
options = buildtimes_df["section_name"].unique()
# labels = [option + ' ' for option in options]

input_dropdown = alt.binding_select(options=options, name="Section Name:")
selection = alt.selection_point(fields=["section_name"], bind=input_dropdown)
```

```python
alt.data_transformers.disable_max_rows()
alt.Chart(buildtimes_df).mark_line().encode(
    x='build_number',
    y='duration_seconds',
).add_params(
    selection
).transform_filter(
    selection
)
```
